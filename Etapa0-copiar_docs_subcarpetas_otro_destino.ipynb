```python
#Importar librerías
import os
import shutil
```

```python
#Definir la ruta general de los archivos (Aquí sub sub carpetas hasta el doc final) y la ruta final
ruta_general = "C:/Users/57350/OneDrive - Agencia Distrital para la Educación Superior, la Ciencia y la Tecnología - Atenea/Soportes de Etnias y registros civiles JU3 y JU4"
ruta_final = "C:/Users/57350/Desktop/PLN-DOC/DOCUMENTOS"
```

```python
#Función para listar las direcciones de los archivos finales en todo ese laberinto de subcarpetas de la ruta general
def listar_rutas_de_archivos(carpeta_raiz):
   rutas_de_archivos = [] 

   for dirpath, dirnames, filenames in os.walk(carpeta_raiz):
       for filename in filenames:
           
           ruta_completa = os.path.join(dirpath, filename)
           rutas_de_archivos.append(ruta_completa)
   return rutas_de_archivos

lista_rutas = list(listar_rutas_de_archivos(ruta_general))

len(lista_rutas)
```

```python
rango = list(range(1,len(lista_rutas)))
```

```python
#definir función para copiar archivos de una capeta a otra.
def copiar_archivos(origen, destino):
    try:
        shutil.copy(origen, destino)
        print(f"Archivo {origen} copiado exitosamente.")
    except Exception as e:
        print(f"No se pudo copiar el archivo {origen}. Error: {e}")

#Aplicar la función
for i in range(1, len(lista_rutas)):
    os.chdir(ruta_general)
    copiar_archivos(lista_rutas[i], ruta_final)
```
